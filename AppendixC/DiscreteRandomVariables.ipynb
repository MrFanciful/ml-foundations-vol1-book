```python
import numpy as np
import matplotlib.pyplot as plt
```

Probability Mass Function

```python
# Plot the PMF of X: number of heads in 3 coin tosses
x_values = np.array([0, 1, 2, 3])
pmf = np.array([1/8, 3/8, 3/8, 1/8])

# Create PMF plot
plt.stem(x_values, pmf, basefmt=" ", linefmt='blue', markerfmt='bo')
plt.xticks(x_values)
plt.yticks(np.linspace(0, 0.5, 6))
plt.xlabel('$x$')
plt.ylabel(r'$p_X(x) = P(X = x)$')
plt.title('PMF of Number of Heads in 3 Coin Tosses')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)

# Annotate bars
for x, y in zip(x_values, pmf):
    plt.text(x, y + 0.02, f'{y:.3f}', ha='center')

plt.savefig('figures/coin_toss_pmf.pdf')
```

Cumulative Distribution Function

```python
x_values = np.array([0, 1, 2, 3])
pmf = np.array([1/8, 3/8, 3/8, 1/8])
cdf = np.cumsum(pmf)

x_plot = np.repeat(x_values, 2)
x_plot = np.concatenate(([-0.5], x_plot[1:], [x_values[-1] + 1]))  # extend for step plot
y_plot = np.repeat(cdf, 2)
y_plot = np.insert(y_plot, 0, 0)

# Plot CDF
plt.plot(x_plot, y_plot, drawstyle='steps-post', color='blue', label='CDF $F_X(x)$')
plt.scatter(x_values, cdf, color='blue')

# Annotate values
for x, y in zip(x_values, cdf):
    plt.text(x, y + 0.05, f'{y:.3f}', ha='center')

plt.title('CDF of Number of Heads in 3 Coin Tosses')
plt.xlabel('$x$')
plt.ylabel(r'$F_X(x) = P(X \leq x)$')
plt.xticks(np.arange(0, 4))
plt.yticks(np.linspace(0, 1, 5))
plt.grid(True, linestyle='--', alpha=0.5)
plt.ylim(0, 1.1)
plt.legend(loc='upper left')
plt.savefig('figures/coin_toss_cdf.pdf')
```
