```python
import numpy as np
import matplotlib.pyplot as plt
```

Probability Density Function

```python
# Define the piecewise PDF
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(1, 2, 100)
y1 = x1
y2 = 2 - x2

# Shade the area from 0.5 to 1.5
x_fill1 = np.linspace(0.5, 1, 100)
y_fill1 = x_fill1
x_fill2 = np.linspace(1, 1.5, 100)
y_fill2 = 2 - x_fill2

plt.plot(x1, y1, label=r'$f(x) = x$', color='blue')
plt.plot(x2, y2, label=r'$f(x) = 2 - x$', color='blue')

plt.fill_between(x_fill1, y_fill1, color='skyblue', alpha=0.5, label=r'$P(0.5 \leq X \leq 1.5)$')
plt.fill_between(x_fill2, y_fill2, color='skyblue', alpha=0.5)

plt.axhline(0, color='black', lw=1)
plt.title('Probability Density Function of a Triangular Distribution')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig('figures/triangular_pdf.pdf')
```
