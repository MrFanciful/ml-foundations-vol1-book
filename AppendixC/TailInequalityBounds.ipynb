```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define range of n values
n_values = np.arange(10, 201, 10)

# Markov bound is constant
markov_bounds = np.full_like(n_values, 2/3, dtype=float)

# Chebyshev bound
chebyshev_bounds = 4 / n_values

# Hoeffding bound
hoeffding_bounds = np.exp(-n_values / 8)

# Chernoff bound
chernoff_bounds = np.exp((n_values / 4) * np.log(16 / 27))

# Plotting
plt.plot(n_values, markov_bounds, label="Markov")
plt.plot(n_values, chebyshev_bounds, label="Chebyshev", linestyle="-.")
plt.plot(n_values, hoeffding_bounds, label="Hoeffding", linestyle=':')
plt.plot(n_values, chernoff_bounds, label="Chernoff", linestyle="--")

plt.yscale('log')
plt.xlabel("$n$")
plt.ylabel("Upper Bound on $P(X \geq 0.75n)$")
plt.title("Comparison of Tail Inequality Bounds")
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()
plt.savefig('figures/tail_inequality_bounds.pdf')
```
