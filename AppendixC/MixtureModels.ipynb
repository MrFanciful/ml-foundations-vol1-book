```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean1, std1 = -1, 1
mean2, std2 = 4, 1.5
w1, w2 = 0.7, 0.3

# Generate points along the x-axis
x = np.linspace(-5, 10, 1000)

# Calculate the individual normal distributions
normal1 = norm.pdf(x, mean1, std1)
normal2 = norm.pdf(x, mean2, std2)

# Calculate the mixture distribution
mixture = w1 * normal1 + w2 * normal2

# Create the plot
plt.plot(x, normal1, '--', label=f'Component 1: $\mu={mean1}, \sigma={std1}$')
plt.plot(x, normal2, '--', label=f'Component 2: $\mu={mean2}, \sigma={std2}$')
plt.plot(x, mixture, '-', color='black', label='Mixture distribution')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/gmm.pdf')
```
