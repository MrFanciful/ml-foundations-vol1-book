```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Correlation values from -1 to 1 in steps of 0.25
correlations = np.round(np.arange(-1.0, 1.01, 0.2), 2)
n = len(correlations)

# Set up grid size
cols = 6
rows = int(np.ceil(n / cols))

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(10, 4))
axes = axes.flatten()

for i, rho in enumerate(correlations):
    # Create bivariate normal data with specified correlation
    mean = [0, 0]
    cov = [[1, rho], [rho, 1]]
    x, y = np.random.multivariate_normal(mean, cov, 100).T

    ax = axes[i]
    ax.scatter(x, y, s=20, edgecolor='deeppink', facecolor='none')
    ax.set_title(f"$\\rho$ = {rho:.1f}", fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])    
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)

# Hide any unused axes
for j in range(i+1, len(axes)):
    axes[j].axis('off')

# Adjusted positions for labels
fig.text(0.52, 0.05, '$x$', ha='center', fontsize=12)
fig.text(0.09, 0.52, '$y$', va='center', rotation='vertical', fontsize=12)

plt.savefig('figures/pearson_correlation.pdf', bbox_inches='tight')
```
