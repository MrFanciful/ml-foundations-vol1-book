```python
import matplotlib.pyplot as plt
import numpy as np
```

Binomial Distribution

```python
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10
p = 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plotting the binomial distribution
plt.vlines(x, 0, pmf, colors='skyblue', lw=3)
plt.plot(x, pmf, 'o', color='darkblue')
plt.title('Binomial Distribution PMF\n$n = 10$, $p = 0.5$')
plt.xlabel('Number of Successes')
plt.ylabel('Probability Mass Function')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.savefig('figures/binomial_distribution.pdf')
```

Poisson Distribution

```python
from scipy.stats import poisson

x = np.arange(0, 21)
lambdas = [1, 4, 10]
colors = ['purple', 'deeppink', 'blue']

plt.figure(figsize=(8, 5))

# Plot Poisson PMFs 
for lam, color in zip(lambdas, colors):
    markerline, stemlines, baseline = plt.stem(x, poisson.pmf(x, lam), basefmt=" ", label=rf'$\lambda = {lam}$')
    plt.setp(markerline, color=color)
    plt.setp(stemlines, color=color)

plt.title('Poisson Distribution PMF for Different Values of $\lambda$')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability $P(X = k)$')
plt.xticks(np.arange(0, 21, 1))
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Rate $\lambda$')
plt.savefig('figures/poisson_distribution.pdf')
```

Categorical Distribution

```python
import matplotlib.pyplot as plt

# Define categories and probability distributions
categories = ['A', 'B', 'C', 'D']
distributions = {
    r'$\mathbf{p} = (0.1, 0.2, 0.5, 0.2)$': [0.1, 0.2, 0.5, 0.2],
    r'$\mathbf{p} = (0.25, 0.25, 0.25, 0.25)$': [0.25, 0.25, 0.25, 0.25],
    r'$\mathbf{p} = (0.05, 0.15, 0.7, 0.1)$': [0.05, 0.15, 0.7, 0.1],
}
colors = ['skyblue', 'lightgreen', 'salmon']

# Plot bar charts for different categorical distributions
fig, axs = plt.subplots(1, 3, figsize=(14, 4), sharey=True)

for ax, (label, probs), color in zip(axs, distributions.items(), colors):
    ax.bar(categories, probs, color=color)
    ax.set_ylim(0, 1)
    ax.set_title(label)
    ax.set_ylabel('Probability')
    ax.grid(True, axis='y', linestyle='--', alpha=0.5)

fig.suptitle('Categorical Distributions with Different Probability Vectors')
plt.savefig('figures/categorical_distribution.pdf', bbox_inches='tight')
```
