```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm

np.random.seed(0)  # for reproducibility
```

```python
mu, sigma = 50, 5  # the normal distribution parameters
n = 100  # number of samples

# Sample data from the distribution
x = np.random.normal(mu, sigma, n)
```

```python
# MLE estimates
mu_ml = x.mean()
sigma_ml = x.std()

print(f'MLE of μ: {mu_ml:.3f}')
print(f'MLE of σ: {sigma_ml:.3f}')
```

```python
# Plot an histogram of the input data
sns.histplot(x, bins=20, stat='density', linewidth=0.5)

# Generate points for the x-axis
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)

dist1 = norm(mu, sigma)
dist2 = norm(mu_ml, sigma_ml)

# Plot the two PDFs   
plt.plot(x, dist1.pdf(x), label='True distribution', color='blue')
plt.plot(x, dist2.pdf(x), label='Estimated distribution', color='green')

plt.xlabel('$x$')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
```
