```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def epsilon_insensitive_loss(y_true, y_pred, epsilon):
    return np.maximum(0, np.abs(y_true - y_pred) - epsilon)
    
def squared_loss(y_true, y_pred):
    return (y_true - y_pred)**2

# Generate a range of predicted values
y_true = 0  # Assume true value is 0 for simplicity
y_pred = np.linspace(-2, 2, 400)  # Predicted values range

# Define different epsilon values to visualize their effect
epsilon_values = [0.1, 0.5, 1.0]

for epsilon in epsilon_values:
    loss = epsilon_insensitive_loss(y_true, y_pred, epsilon)
    plt.plot(y_pred, loss, label=f'$\epsilon = {epsilon}$')
    
# Plotting squared loss
squared_loss_values = squared_loss(y_true, y_pred)
plt.plot(y_pred, squared_loss_values, label='Squared Loss', linestyle='--', color='deeppink')

plt.xlabel('$y - \hat{y}$')
plt.ylabel('$\epsilon$-Insensitive Loss')
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/epsilon_insensitive_loss.pdf')
```
