```python
import numpy as np
import matplotlib.pyplot as plt

# Define the RBF kernel function
def rbf_kernel(distance, gamma):
    return np.exp(-gamma * distance**2)

# Generate distances from 0 to 2 in 100 steps
distances = np.linspace(0, 2, 100)

# Define different gamma values to demonstrate their effect
gamma_values = [0.5, 1, 2]  # Lower, medium, and higher sensitivity

# Plot RBF kernel values for each gamma on the same graph
for gamma in gamma_values:
    kernel_values = rbf_kernel(distances, gamma)
    plt.plot(distances, kernel_values, label=f'$\gamma = {gamma}$')

plt.xlabel('Distance between points')
plt.ylabel('Kernel output')
plt.legend()
plt.grid(alpha=0.5)

plt.savefig('figures/rbf_kernel.pdf')
```
