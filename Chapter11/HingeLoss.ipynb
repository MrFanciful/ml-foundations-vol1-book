```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-4, 4, 400)

zero_one_loss = np.where(x < 0, 1, 0)
hinge_loss = np.maximum(0, 1 - x)
log_loss = np.log2(1 + np.exp(-x))
```

```python
plt.plot(x, hinge_loss, label='Hinge Loss', color='tab:blue')
plt.plot(x, log_loss, label='Log Loss', linestyle='--', color='tab:green')
plt.plot(x, zero_one_loss, label='Zero-One Loss', linestyle=':', color='tab:red')
plt.xlabel('Decision Function Output $y \hat{y}$')
plt.ylabel('Loss')
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/hinge_loss.pdf')
```
