```python
import numpy as np
import matplotlib.pyplot as plt
```

Hinge Loss

```python
x = np.linspace(-4, 4, 400)

zero_one_loss = np.where(x < 0, 1, 0)
hinge_loss = np.maximum(0, 1 - x)
log_loss = np.log2(1 + np.exp(-x))
```

```python
plt.plot(x, zero_one_loss, label='Zero-One Loss', linestyle='--')
plt.plot(x, hinge_loss, label='Hinge Loss')
plt.plot(x, log_loss, label='Log Loss', linestyle='-.')
plt.xlabel('Decision Function Output $y \hat{y}$')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.savefig('figures/hinge_loss.pdf', bbox_inches='tight')
```

Epsilon-Insensitive Loss

```python
def epsilon_insensitive_loss(y_true, y_pred, epsilon):
    return np.maximum(0, np.abs(y_true - y_pred) - epsilon)
    
def squared_loss(y_true, y_pred):
    return (y_true - y_pred)**2

# Generate a range of predicted values
y_true = 0  # Assume true value is 0 for simplicity
y_pred = np.linspace(-2, 2, 400)  # Predicted values range

# Define different epsilon values to visualize their effect
epsilon_values = [0.1, 0.5, 1.0]

for epsilon in epsilon_values:
    loss = epsilon_insensitive_loss(y_true, y_pred, epsilon)
    plt.plot(y_pred, loss, label=f'$\epsilon = {epsilon}$')
    
# Plotting squared loss
squared_loss_values = squared_loss(y_true, y_pred)
plt.plot(y_pred, squared_loss_values, label='Squared Loss', linestyle='--', color='gray')

plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.xlabel('Predicted Value')
plt.ylabel('$\epsilon$-Insensitive Loss')
plt.legend()
plt.grid(True)
plt.savefig('figures/epsilon_insensitive_loss.pdf', bbox_inches='tight')
```
