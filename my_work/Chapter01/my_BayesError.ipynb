```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import norm
from scipy.optimize import fsolve
```

```python
mu_coyote, sigma_coyote = 60, 10
mu_wolf, sigma_wolf = 80, 8
```

```python
# Generate points on the x axis (heights) from 20 to 120 cm
x = np.linspace(20, 120, 1000)

#Probability density function (PDFs) for the two species
dist_coyote = norm(mu_coyote, sigma_coyote)
dist_wolf = norm(mu_wolf, sigma_wolf)

# Plot the PDFs
plt.plot(x, dist_coyote.pdf(x), label='Coyote', color='blue')
plt.fill_between(x, dist_coyote.pdf(x), color='blue', alpha=0.3)
plt.plot(x, dist_wolf.pdf(x), label='Wolf', color='green')
plt.fill_between(x, dist_wolf.pdf(x), color='green', alpha=0.3)

plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.legend()
plt.savefig('figures/bayes_error.png')
```

```python
def find_intersection(x):
    return dist_coyote.pdf(x) - dist_wolf.pdf(x)

intersection_point = fsolve(find_intersection, 70)
print(f'Intersection point: {intersection_point[0]:.3f}')
```
