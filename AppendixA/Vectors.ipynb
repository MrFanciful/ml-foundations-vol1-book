```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the vector in 2D
vector = np.array([5, 3])

fig, ax = plt.subplots()

# Plot the vector as an arrow starting at the origin
ax.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='b', linewidth=1)

# Plot the point at the vector's tip
ax.scatter(vector[0], vector[1], color='r', s=100)

# Add text to display the coordinates 
ax.text(vector[0], vector[1] + 0.5, f'({vector[0]}, {vector[1]})', color='black', ha='center', 
        fontsize=10, bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_xlim(-1, vector[0] + 2)
ax.set_ylim(-1, vector[1] + 2)
ax.set_aspect('equal', 'box')
ax.grid(alpha=0.5)

plt.savefig('figures/vector.pdf')
```

Vector Subtraction

```python
# Define vectors 
x = np.array([4, 3])
y = np.array([2, 5])
negative_y = -y
difference = x - y

fig, ax = plt.subplots()

# Left side: Original subtraction
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='b', linewidth=2, label=r'$\vec{x}$')
ax.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='g', linewidth=2, label=r'$\vec{y}$')
ax.quiver(y[0], y[1], difference[0], difference[1], angles='xy', scale_units='xy', 
          scale=1, color='r', linewidth=2, label=r'$\vec{x} - \vec{y}$')

# Add points and labels for the left diagram
ax.scatter([x[0], y[0]], [x[1], y[1]], color='black', s=50)
ax.text(x[0], x[1] + 0.3, r'$\vec{x}$', color='b', ha='center', fontsize=12)
ax.text(y[0], y[1] + 0.3, r'$\vec{y}$', color='g', ha='center', fontsize=12)
ax.text(difference[0] + y[0] / 2 + 0.5, difference[1] + y[1] / 2 + 3.8, r'$\vec{x} - \vec{y}$', 
        color='r', ha='center', fontsize=12)

# Right side: Addition of negative vector
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='b', linewidth=2)
ax.quiver(0, 0, negative_y[0], negative_y[1], angles='xy', scale_units='xy', scale=1, 
          color='orange', linewidth=2, label=r'$-\vec{y}$')
ax.quiver(0, 0, difference[0], difference[1], angles='xy', scale_units='xy', scale=1, 
          color='r', linewidth=2, label=r'$\vec{x} + (-\vec{y})$')

# Add points and labels for the right diagram
ax.scatter([x[0], negative_y[0]], [x[1], negative_y[1]], color='black', s=50)
ax.text(negative_y[0], negative_y[1] - 0.5, r'$-\vec{y}$', color='orange', ha='center', fontsize=12)
ax.text(difference[0] / 2 + 0.75, difference[1] / 2 + 0.3, r'$\vec{x} + (-\vec{y})$', 
        color='r', ha='center', fontsize=12)

ax.axhline(0, color='gray', linewidth=0.8)
ax.axvline(0, color='gray', linewidth=0.8)
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.grid(alpha=0.5)
ax.set_aspect('equal', 'box')

plt.savefig('figures/vector_subtraction.pdf')
```

Angle Between Vectors

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
vec_x = np.array([3, 3])
vec_y = np.array([5, 0])

# Compute dot product and norms
dot_product = np.dot(vec_x, vec_y)
norm_x = np.linalg.norm(vec_x)
norm_y = np.linalg.norm(vec_y)
cos_theta = dot_product / (norm_x * norm_y)

# Angle in degrees
angle_deg = np.degrees(np.arccos(cos_theta))

# Plot the vectors
fig, ax = plt.subplots()
ax.quiver(0, 0, vec_x[0], vec_x[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\vec{x}$ = (3, 3)')
ax.quiver(0, 0, vec_y[0], vec_y[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\vec{y}$ = (5, 0)')

ax.text(vec_x[0] + 0.05, vec_x[1] + 0.05, '(3, 3)', fontsize=10, color='r')
ax.text(vec_y[0] + 0.05, vec_y[1] + 0.05, '(5, 0)', fontsize=10, color='b')

# Annotate the angle and computation details
ax.text(0.5, 0.2, r'$\theta = 45^\circ$', fontsize=14, color='purple')

ax.set_xlim(-1, 6)
ax.set_ylim(-1, 4)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')

ax.grid(alpha=0.5)
plt.savefig('figures/angle_between_vectors.pdf')
```

Projections

```python
# Define vectors
x = np.array([3, 2])
y = np.array([4, 1])

# Compute scalar projection
scalar_proj = np.dot(x, y) / np.linalg.norm(y)

# Compute vector projection
vector_proj = (np.dot(x, y) / np.dot(y, y)) * y

# Create figure
fig, ax = plt.subplots()
ax.set_xlim(0, 5)
ax.set_ylim(0, 3)

# Plot vectors
ax.quiver(0, 0, x[0], x[1], angles='xy', scale_units='xy', scale=1, color='r', label=r'$\vec{x}$')
ax.quiver(0, 0, y[0], y[1], angles='xy', scale_units='xy', scale=1, color='b', label=r'$\vec{y}$')
ax.quiver(0, 0, vector_proj[0], vector_proj[1], angles='xy', scale_units='xy', scale=1, 
          color='g', label=r'$\operatorname{proj}_{\vec{y}} \vec{x}$')

# Plot projection line
ax.plot([x[0], vector_proj[0]], [x[1], vector_proj[1]], 'k--', alpha=0.8)

# Labels for vectors with coordinates
ax.text(x[0], x[1], r'$\vec{x}$' + f' ({x[0]}, {x[1]})', fontsize=12, verticalalignment='bottom', 
        horizontalalignment='right', color='r')
ax.text(y[0], y[1], r'$\vec{y}$' + f' ({y[0]}, {y[1]})', fontsize=12, verticalalignment='bottom', 
        horizontalalignment='left', color='b')
ax.text(vector_proj[0] + 1, vector_proj[1] - 0.2, r'$\operatorname{proj}_{\vec{y}} \vec{x}$' + 
        f' ({vector_proj[0]:.2f}, {vector_proj[1]:.2f})', fontsize=12, 
        verticalalignment='top', horizontalalignment='right', color='g')

ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.grid(alpha=0.5)
ax.legend()

plt.savefig('figures/vector_projection.pdf')
```
