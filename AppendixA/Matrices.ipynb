```python
import numpy as np
import matplotlib.pyplot as plt
```

Eigenvectors

```python
# Define matrix transformation parameters
x_val, y_val = 1, 1
lambda_val = 1.5

# Define vectors
original_vector = np.array([x_val, y_val])
scaled_vector = lambda_val * original_vector

plt.figure(figsize=(5, 5))

# Plot vectors
plt.quiver(0, 0, original_vector[0], original_vector[1], angles='xy', scale_units='xy', 
           scale=1, color='red', width=0.01, label=r'$\vec{x}$')

plt.quiver(original_vector[0] - 0.01, original_vector[1] - 0.01,
           scaled_vector[0] - original_vector[0], scaled_vector[1] - original_vector[1],
           angles='xy', scale_units='xy', scale=1, color='red', alpha=0.75, width=0.008)

# Add dashed projection lines
plt.plot([original_vector[0], original_vector[0]], [0, original_vector[1]], linestyle='dotted', color='gray')
plt.plot([scaled_vector[0], scaled_vector[0]], [0, scaled_vector[1]], linestyle='dotted', color='gray')
plt.plot([0, original_vector[0]], [original_vector[1], original_vector[1]], linestyle='dotted', color='gray')
plt.plot([0, scaled_vector[0]], [scaled_vector[1], scaled_vector[1]], linestyle='dotted', color='gray')

# Annotate vectors and projections
plt.text(original_vector[0] - 0.3, original_vector[1] - 0.15, r'$\vec{x}$', fontsize=12, color='red')
plt.text(scaled_vector[0] - 0.1, scaled_vector[1] + 0.05, r'$A\vec{x} = \lambda \vec{x}$', fontsize=12, color='red', alpha=0.75)
plt.text(x_val + 0.05, 0.03, r'$x$', fontsize=12, color='black')
plt.text(0.05, y_val + 0.05, r'$y$', fontsize=12, color='black')
plt.text(lambda_val * x_val + 0.05, 0.03, r'$\lambda x$', fontsize=12, color='black')
plt.text(0.05, lambda_val * y_val + 0.05, r'$\lambda y$', fontsize=12, color='black')

plt.xlim(0, lambda_val * x_val + 0.5)
plt.ylim(0, lambda_val * y_val + 0.5)
plt.grid(True, alpha=0.5)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.savefig('figures/eigenvector.pdf')
```

Determinant

```python
# Define a 2x2 transformation matrix A
A = np.array([[2, 1],
              [1, 1]])
detA = np.linalg.det(A)

# Define the unit square (closed loop)
unit_square = np.array([[0, 0],
                        [1, 0],
                        [1, 1],
                        [0, 1],
                        [0, 0]])

# Apply the transformation to the unit square
transformed_square = unit_square @ A.T  # Use A.T since we're transforming row vectors

# Compute images of the standard basis vectors
e1 = np.array([1, 0])
e2 = np.array([0, 1])
Ae1 = A @ e1
Ae2 = A @ e2

# Create the plot
plt.figure(figsize=(8, 8))
plt.plot(unit_square[:, 0], unit_square[:, 1], 'b-', label='Unit Square')
plt.plot(transformed_square[:, 0], transformed_square[:, 1],
         'r-', linewidth=2, label='Transformed Parallelogram')

# Draw arrows for the transformed basis vectors
origin = np.array([0, 0])
plt.arrow(origin[0], origin[1], Ae1[0], Ae1[1],
          head_width=0.1, head_length=0.1, fc='g', ec='g', label='A·[1,0]')
plt.arrow(origin[0], origin[1], Ae2[0], Ae2[1],
          head_width=0.1, head_length=0.1, fc='m', ec='m', label='A·[0,1]')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(alpha=0.5)
plt.gca().set_aspect('equal', adjustable='box')

plt.savefig('figures/determinant.pdf', bbox_inches='tight')
```
