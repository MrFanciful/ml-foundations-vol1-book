```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x, y) = x^2 + y^2
def f(x, y):
    return x**2 + y**2

# Generate data for the plots
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Surface Plot 
fig1 = plt.figure(figsize=(6, 5))  
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(X, Y, Z, cmap="Blues", edgecolor='k', alpha=0.7)
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_zlabel('$f(x, y)$')
ax1.set_title('Surface Plot of $f(x, y) = x^2 + y^2$')

plt.tight_layout()
plt.savefig('figures/level_sets_surface.pdf', bbox_inches='tight')  

# Contour Plot
fig2 = plt.figure(figsize=(6, 5))  
ax2 = fig2.add_subplot(111)
contours = ax2.contour(X, Y, Z, levels=10, cmap="plasma")
ax2.clabel(contours, inline=True, fontsize=8)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_title('Contour Plot of $f(x, y) = x^2 + y^2$')

plt.tight_layout()
plt.savefig('figures/level_sets_contour.pdf', bbox_inches='tight')  
```
