```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function and Taylor polynomials
x = np.linspace(-2.5, 2.5, 200)  # Narrowed domain
true_exp = np.exp(x)
P2 = 1 + x + (x**2)/2
P3 = P2 + (x**3)/6
P4 = P3 + (x**4)/24

# Plot the functions
plt.plot(x, true_exp, label=r"$e^x$", color="black")
plt.plot(x, P2, label=r"$P_2(x)$", linestyle="dotted", color='red')
plt.plot(x, P3, label=r"$P_3(x)$", linestyle="dashdot", color='green')
plt.plot(x, P4, label=r"$P_4(x)$", linestyle="dashed", color='blue')

plt.xlabel(r"$x$")
plt.ylabel(r"$f(x)$")
plt.title(r"Comparison of $e^x$ and Taylor Polynomials")
plt.legend()
plt.grid(alpha=0.5)

plt.savefig('figures/taylor_polynomials.pdf')
```
