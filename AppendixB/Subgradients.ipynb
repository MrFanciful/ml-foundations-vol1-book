```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the absolute value function
x = np.linspace(-2, 2, 400)
y = np.abs(x)

# Define supporting lines for subgradients in [-1, 1]
slopes = [-1, -0.5, 0, 0.5, 1]  
x0 = 0
y0 = np.abs(x0)

# Create the plot
plt.plot(x, y, label=r'$f(x) = |x|$', color='black', linewidth=3)

# Plot supporting lines at x=0 for different subgradients
for m in slopes:
    plt.plot(x, y0 + m * (x - x0), linestyle='dashed', label=f'$c={m}$')

# Highlight the point of non-differentiability
plt.scatter([x0], [y0], color='red', zorder=3)

plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend(loc='upper center', prop={'size': 9})
plt.grid(True, alpha=0.5)

plt.savefig('figures/subgradients.pdf')
```
