```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define functions for different types of non-differentiability
x = np.linspace(-1, 1, 400)

# Absolute value function (Corner)
f1 = np.abs(x)

# Cusp function: |x|^(2/3)
f2 = np.abs(x) ** (2/3)

# Vertical tangent: x^(1/3)
f3 = np.sign(x) * np.abs(x) ** (1/3)

# Weierstrass function (approximation using 10 terms)
def weierstrass(x, a=0.5, b=3, n_terms=10):
    return sum(a**n * np.cos(b**n * np.pi * x) for n in range(n_terms))

f4 = weierstrass(x)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot each function
axs[0, 0].plot(x, f1, 'b')
axs[0, 0].set_title(r"$f(x) = |x|$ (Corner at $x=0$)")

axs[0, 1].plot(x, f2, 'g')
axs[0, 1].set_title(r"$f(x) = |x|^{2/3}$ (Cusp at $x=0$)")

axs[1, 0].plot(x, f3, 'r')
axs[1, 0].set_title(r"$f(x) = x^{1/3}$ (Vertical Tangent at $x=0$)")

axs[1, 1].plot(x, f4, 'm')
axs[1, 1].set_title("Weierstrass Function (Nowhere Differentiable)")

for ax in axs.flat:
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$f(x)$')

plt.tight_layout()
plt.savefig('figures/nondifferentiability.pdf', bbox_inches='tight')
```
