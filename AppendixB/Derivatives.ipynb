```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the function
def f(x):
    return x**2

# Derivative of the function
def df(x):
    return 2*x

# Points for the secant line
x1, x2 = 2, 6
y1, y2 = f(x1), f(x2)

# Compute secant line slope
secant_slope = (y2 - y1) / (x2 - x1)
secant_line = lambda x: secant_slope * (x - x1) + y1

# Tangent line at x1
tangent_slope = df(x1)
tangent_line = lambda x: tangent_slope * (x - x1) + y1

# Generate x values
x_vals = np.linspace(-2, 8, 400)
y_vals = f(x_vals)

# Create plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x_vals, y_vals, 'k', label=r'$f(x) = x^2$')  # Function curve
ax.plot(x_vals, secant_line(x_vals), 'b--', label='Secant Line')  # Secant line
ax.plot(x_vals, tangent_line(x_vals), 'r', label='Tangent Line')  # Tangent line

# Mark points
ax.scatter([x1, x2], [y1, y2], color='blue', zorder=3)
ax.text(x1 - 0.5, y1 - 5, r'$(x, f(x))$', fontsize=12, verticalalignment='top')
ax.text(x2 - 2, y2 + 2.5, r'$(x+h, f(x+h))$', fontsize=12, verticalalignment='bottom')

# Labels and title
ax.set_xlabel(r'$x$', fontsize=14)
ax.set_ylabel(r'$f(x)$', fontsize=14)
ax.set_title("Derivative as the Limit of Secant Lines", fontsize=14)
ax.legend()
ax.grid(True, alpha=0.5)

plt.savefig('figures/derivative.pdf')
```
