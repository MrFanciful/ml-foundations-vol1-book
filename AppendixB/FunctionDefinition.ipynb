```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x_values = np.linspace(-10, 10, 400)
y_values = sigmoid(x_values)

# Create the plot
plt.figure()
plt.plot(x_values, y_values, label=r"$f(x) = \dfrac{1}{1 + e^{-x}}$", linewidth=2)

# Highlight domain, range, and asymptotes
plt.axhline(y=1, color='red', linestyle='dashed', linewidth=1, label="Upper bound: $y=1$")
plt.axhline(y=0, color='purple', linestyle='dashed', linewidth=1, label="Lower bound: $y=0$")

plt.xlabel("$x$", fontsize=12)
plt.ylabel("$f(x)$", fontsize=12)
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/sigmoid.pdf')
```
