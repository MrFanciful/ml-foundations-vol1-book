```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4 - 2*x**2  # A quartic function with local extrema

# Generate x values
x_vals = np.linspace(-2, 2, 400)
y_vals = f(x_vals)

# Find the function values at critical points
x_crit_points = [-1, 0, 1]  # Solving f'(x) = 4x^3 - 4x = 0 gives x = -1, 0, 1
y_crit_points = [f(x) for x in x_crit_points]

# Create the plot
plt.plot(x_vals, y_vals, label=r"$f(x) = x^4 - 2x^2$", color="blue")

# Mark the local minima (x = -1, x = 1) and the saddle point (x = 0) 
plt.scatter([x_crit_points[0], x_crit_points[2]], [y_crit_points[0], y_crit_points[2]], 
            color="red", label="Global minima ($x = \pm 1$)")
plt.scatter([x_crit_points[1]], [y_crit_points[1]], 
            color="green", label="Local maximum ($x=0$)")

plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")  # y-axis
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/extremum_points.pdf')
```
