```python
import numpy as np
import matplotlib.pyplot as plt
```

Average side length of a k-neighborhood

```python
k = 10
n = 10**6
d = np.arange(1, 501)
l = (k/n)**(1/d)

plt.plot(d, l)
plt.xlabel('Number of dimensions')
plt.ylabel('Side length of the neighborhood')

plt.savefig('figures/average_neighborhood_length.pdf')
```
