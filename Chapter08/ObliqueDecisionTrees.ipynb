```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

```python
# Number of samples
n_samples = 100

# Generate random points uniformly distributed within [0, 1]
x = np.random.rand(n_samples)
y = np.random.rand(n_samples)

# Classify points
labels = np.where(y < x, 0, 1)  # Class 0 if y < x, else Class 1
```

```python
# Plotting
plt.figure(figsize=(8, 6))
plt.scatter(x[labels == 0], y[labels == 0], color='r', label='Class 0')
plt.scatter(x[labels == 1], y[labels == 1], color='b', label='Class 1')
plt.plot([0, 1], [0, 1], '--', color='m', label='Decision boundary')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True)
plt.axis([0, 1, 0, 1])
plt.savefig('figures/oblique_decision_trees.pdf')
```
