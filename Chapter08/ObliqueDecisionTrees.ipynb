```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_positive = [2.7, 3.2, 3.7, 4.0, 5.2, 5.3, 5.8, 6.2, 6.9, 7.2]
y_positive = [6.0, 5.2, 5.4, 4.8, 4.6, 3.5, 2.5, 3.8, 2.4, 1.5]
x_negative = [0.7, 1.0, 1.5, 2.0, 2.2, 2.5, 3.3, 3.6, 4.0, 4.5, 5.8, 6.5]
y_negative = [5.5, 4.2, 4.5, 2.8, 3.8, 2.2, 3.7, 1.2, 2.5, 2.8, 1.8, 0.5]

# Plot data points for positive and negative classes
plt.scatter(x_positive, y_positive, marker='+', color='black', s=50)
plt.scatter(x_negative, y_negative, marker='_', color='black', s=50)

# Plot staircase axis-aligned boundaries 
plt.plot([2.5, 2.5], [4.0, 8], 'k-', linewidth=1, label='Univariate Splits') 
plt.plot([2.5, 5.0], [4.0, 4.0], 'k-', linewidth=1)
plt.plot([5.0, 5.0], [4.0, 2.0], 'k-', linewidth=1)
plt.plot([5.0, 7.0], [2.0, 2.0], 'k-', linewidth=1)
plt.plot([7.0, 7.0], [2.0, 0.0], 'k-', linewidth=1)

# Oblique boundary
x_vals = np.array([0.0, 8.0])
y_vals = 8.0 - x_vals 
plt.plot(x_vals, y_vals, '--', color='blue', label='Oblique Boundary')

# Set labels and title
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.xlim([0, 8])
plt.ylim([0, 8])
plt.legend()
plt.savefig('figures/oblique_decision_tree.pdf')
```
