```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def entropy(p):
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)
```

```python
def gini(p):
    return p * (1 - p) + (1 - p) * (1 - (1 - p))
```

```python
x = np.linspace(0, 1, 100)
ent = [entropy(p) if p != 0 and p != 1 else 0 for p in x]
scaled_ent = [e * 0.5 for e in ent]
```

```python
plt.plot(x, scaled_ent, label='Entropy (scaled)')
plt.plot(x, gini(x), label='Gini index')

plt.xlabel('$p$')
plt.ylabel('Impurity value')
plt.legend()
plt.savefig('figures/node_impurity_measures.pdf')
```
