```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from scipy.stats import gamma

# Parameters
alpha_prior = 3
beta_prior = 1

alpha_post = alpha_prior + 1  # One observation x = 2
beta_post = beta_prior + 2

theta = np.linspace(0.001, 6, 500)
prior_pdf = gamma.pdf(theta, a=alpha_prior, scale=1/beta_prior)
post_pdf = gamma.pdf(theta, a=alpha_post, scale=1/beta_post)

# Plotting
plt.plot(theta, prior_pdf, label='Prior: Gamma(3, 1)', linewidth=2)
plt.plot(theta, post_pdf, label='Posterior: Gamma(4, 3)', linewidth=2)
plt.axvline(x=1, color='red', linestyle='--', label='MAP = 1')
plt.title("Prior and Posterior Distributions for $\\theta$")
plt.xlabel(r"Rate parameter $\theta$")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.5)
plt.tight_layout()
plt.savefig('figures/maximum_a_posteriori.pdf')
```
