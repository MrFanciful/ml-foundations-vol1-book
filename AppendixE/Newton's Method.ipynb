```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x - 2)**2 + 0.05 * (x - 2)**3

def grad_f(x):
    return 2 * (x - 2) + 0.15 * (x - 2)**2

def hess_f(x):
    return 2 + 0.3 * (x - 2)

# Initial guess
x0 = 0.5
num_steps = 15

# Gradient descent
x_gd = [x0]
alpha = 0.1
for _ in range(num_steps):
    x_new = x_gd[-1] - alpha * grad_f(x_gd[-1])
    x_gd.append(x_new)

# Newton's method
x_newton = [x0]
for _ in range(num_steps):
    g = grad_f(x_newton[-1])
    h = hess_f(x_newton[-1])
    x_new = x_newton[-1] - g / h
    x_newton.append(x_new)

# Plot the function and both methods
x_vals = np.linspace(-1, 4, 500)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, 'k-', label='Objective function')
plt.plot(x_gd, [f(x) for x in x_gd], 'o--', color='purple', label='Gradient descent')
plt.plot(x_newton, [f(x) for x in x_newton], 'o--', color='blue', label="Newton's method")
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.grid(alpha=0.5)
plt.savefig('figures/newton_vs_gradient_descent.pdf')
```
