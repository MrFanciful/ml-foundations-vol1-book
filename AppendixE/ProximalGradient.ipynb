```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a 1D non-smooth convex function: f(x) = |x|
def h(x):
    return np.abs(x)

# Define the proximal operator of f for lambda = 1 (soft-thresholding)
def prox_h(v, lam=1.0):
    return np.sign(v) * np.maximum(np.abs(v) - lam, 0)

# Generate a range of v values
v = np.linspace(-4, 4, 400)
prox_vals = prox_h(v, lam=1.0)

# Plot
plt.plot(v, v, 'r--', label=r'Identity: $v$')  # Diagonal line
plt.plot(v, prox_vals, 'b-', label=r'$\mathrm{prox}_{\lambda f}(v)$ with $f(x)=|x|$, $\lambda=1$')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel(r'$v$')
plt.ylabel(r'Proximal value')
plt.legend()
plt.grid(alpha=0.5)
plt.savefig('figures/proximal_operator.pdf')
```
