```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set up the x-axis range
x = np.linspace(-1.2, 1.2, 300)

# Define functions and tangents
# Negative curvature: f(x) = -x^2 + 1
f_neg = lambda x: -x**2 + 1
x0_neg = 0.3
y0_neg = f_neg(x0_neg)
slope_neg = -2 * x0_neg
tangent_neg = y0_neg + slope_neg * (x - x0_neg)

# No curvature: f(x) = -0.5x + 0.5 (linear)
f_lin = lambda x: -0.5 * x + 0.5

# Positive curvature: f(x) = x^2
f_pos = lambda x: x**2
x0_pos = -0.3
y0_pos = f_pos(x0_pos)
slope_pos = 2 * x0_pos
tangent_pos = y0_pos + slope_pos * (x - x0_pos)

# Set up subplots
fig, axes = plt.subplots(1, 3, figsize=(12, 3), sharey=True)

# Plot: Negative curvature
axes[0].plot(x, f_neg(x), 'b', label='$f(x)$')
axes[0].plot(x, tangent_neg, 'r--', label='Tangent')
axes[0].set_title("Negative curvature", fontsize=14)
axes[0].set_xlabel("$x$", fontsize=12)
axes[0].set_ylabel("$f(x)$", fontsize=12)
axes[0].set_xlim([-1.2, 1.2])
axes[0].set_ylim([-1, 2])
axes[0].grid(alpha=0.5)
axes[0].legend(fontsize=11)

# Plot: No curvature
axes[1].plot(x, f_lin(x), 'b', label='$f(x)$')
axes[1].plot(x, f_lin(x), 'r--', label='Tangent')
axes[1].set_title("No curvature", fontsize=14)
axes[1].set_xlabel("$x$", fontsize=12)
axes[1].set_xlim([-1.2, 1.2])
axes[1].set_ylim([-1, 2])
axes[1].grid(alpha=0.5)
axes[1].legend(fontsize=11)

# Plot: Positive curvature
axes[2].plot(x, f_pos(x), 'b', label='$f(x)$')
axes[2].plot(x, tangent_pos, 'r--', label='Tangent')
axes[2].set_title("Positive curvature", fontsize=14)
axes[2].set_xlabel("$x$", fontsize=12)
axes[2].set_xlim([-1.2, 1.2])
axes[2].set_ylim([-1, 2])
axes[2].grid(alpha=0.5)
axes[2].legend(fontsize=11)

plt.savefig('figures/curvature_effect.pdf', bbox_inches='tight')
```
