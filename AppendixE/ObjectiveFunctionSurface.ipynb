```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the function with multiple local minima/maxima, flat regions, and saddle points
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) + 0.1 * (x**2 - y**2)

# Generate grid points for the plot
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create a 3D plot
fig = plt.figure(figsize=(10, 7))
#fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='Blues', alpha=0.8)

# Labels and title
ax.set_xlabel('$w_1$', fontsize=14)
ax.set_ylabel('$w_2$', fontsize=14)
ax.set_zlabel('$J(w_1, w_2)$', fontsize=14)

# Save and show the plot
plt.tight_layout()
plt.savefig('figures/objective_function_surface.pdf')
```
