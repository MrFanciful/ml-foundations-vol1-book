```python
import numpy as np
import matplotlib.pyplot as plt
```

Perceptron Activation Functions

```python
# Define the step function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Define the sign function
def sign_function(x):
    return np.where(x >= 0, 1, -1)

# Generate input values
x = np.linspace(-3, 3, 400)

# Compute the outputs of the step and sign functions
y_step = step_function(x)
y_sign = sign_function(x)

# Plot the step and sign functions on the same figure
# Step function
plt.plot(x, y_step, label='Step function', linestyle='-', color='blue')

# Sign function
plt.plot(x, y_sign, label='Sign function', linestyle='--', color='red')

# Add labels, legend, and grid
plt.axhline(0, color='black', linewidth=0.8, linestyle=':')
plt.axvline(0, color='black', linewidth=0.8, linestyle=':')
plt.xlabel("Input (z)")
plt.ylabel("Activation Value")
plt.yticks([-1, 0, 1], ['-1', '0', '1'])
plt.legend()
plt.grid(alpha=0.5)

plt.savefig('figures/perceptron_activation_functions.pdf')
```

```python
z = np.linspace(-3, 3, 500)
sigmoid = 1 / (1 + np.exp(-z))
tanh = np.tanh(z)
relu = np.maximum(0, z)

# Sigmoid
plt.plot(z, sigmoid, label='Sigmoid ($\\sigma(z)$)')

# Tanh
plt.plot(z, tanh, label='Tanh ($\\tanh(z)$)')

# ReLU
plt.plot(z, relu, label='ReLU ($\\max(0, z)$)')

plt.xlabel('$z$')
plt.ylabel('Activation Value')
plt.axhline(0, color='black', linewidth=0.5, linestyle=':')  # Add a dashed line for y=0
plt.axvline(0, color='black', linewidth=0.5, linestyle=':')  # Add a dashed line for x=0
plt.legend(fontsize=10, loc='upper left')
plt.grid(alpha=0.3)

# Show the plot
plt.savefig('figures/mlp_activation_functions.pdf')
```
